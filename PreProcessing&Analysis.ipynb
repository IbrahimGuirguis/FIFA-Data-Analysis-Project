```python
import numpy as np

# Assuming 'data' is your DataFrame and 'column_name' is the name of the column you want to analyze
column_data = df["value_eur"]

# Calculate the IQR (interquartile range)
Q1 = np.percentile(column_data, 25)
Q3 = np.percentile(column_data, 75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

# Find the outliers in the column
outliers = column_data[(column_data < lower_fence) | (column_data > upper_fence)]

# Print the outliers
sorted_outliers = outliers.sort_values(ascending=True)
print(sorted_outliers)
```

```python
# Count the number of outliers above the upper fence
outliers_above = column_data[column_data > upper_fence].count()

# Count the number of outliers below the lower fence
outliers_below = column_data[column_data < lower_fence].count()
print(outliers_below/len(column_data) ,outliers_above/len(column_data))
```

```python
import numpy as np

# Assuming 'data' is your DataFrame and 'column_name' is the name of the column you want to winsorize
column_data = copy["value_eur"]

winsor_percent = 15  # Set the winsorization percentile
lower_limit = np.percentile(column_data, winsor_percent)  # Calculate the lower winsorization limit
upper_limit = np.percentile(column_data, 100 - winsor_percent)  # Calculate the upper winsorization limit

# Apply winsorization using np.clip
column_data_winsorized = np.clip(column_data, lower_limit, upper_limit)

# Assign the winsorized column back to the DataFrame
copy["value_eur"] = column_data_winsorized
```

```python
#Import Libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
data = pd.read_csv('selected_columns.csv')
target = data['overall']
features = data.drop('overall', axis=1)
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.33, random_state=44, shuffle =True)
#Applying Linear Regression Model 

LinearRegressionModel = LinearRegression()
LinearRegressionModel.fit(X_train, y_train)

print('Linear Regression Train Score is : ' , LinearRegressionModel.score(X_train, y_train))
print('Linear Regression Test Score is : ' , LinearRegressionModel.score(X_test, y_test))
```
